```python
import pandas as pd
import numpy as np
import win32com.client as win32
import datetime
from datetime import date
```

```python
#Visualizar o arquivo selecionado
tabela = pd.read_excel(r"C:\Users\karin\OneDrive\Documentos\Meus Projetos\Pyhton\Total_Faturamento.xlsx")
display(tabela)
```

```python
#Verificar a quantidade de linhas e colunas, respectivamente:
tabela.shape
```

```python
#Alterar o nome de uma coluna da tabela:
tabela.rename(columns = {'Nota Fiscal' : 'NFe'}, inplace = True)
display(tabela)
```

```python
#Remover a última linha com a soma total para não duplicar o valor
tabela = tabela[:-1]
display(tabela)
```

```python
#Calcular o total Geral 
Total = tabela["Total"].sum()
print(Total)
```

```python
#Verificar os tipos de dados: 
tabela.dtypes
```

```python
#Alterar o tipo de dado da NFe de float64 para tipo int:
tabela.NFe = tabela.NFe.astype(int)
tabela.dtypes
```

```python
#Verificar a tabela após a alteração do tipo de dado:
display(tabela)
```

```python
#Verificar dados estatísticos da pivot table
tabela.Total.describe()
```

```python
#Criar a tabela dinâmica (Nome do Cliente x Total)
df_pivot = pd.pivot_table(tabela, values = "Total", index = "Nome do Cliente", aggfunc = 'sum', margins = True, margins_name='Total Geral')
display(df_pivot)
```
